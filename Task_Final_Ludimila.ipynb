## Agrupamento de Dados e Plotagem de Gráficos:



a) Quantidade de Funcionários por Departamento:

```python
df_qtde_funcionarios = df.groupby("Department")[["EmployeeCount"]].count()
df_qtde_funcionarios
```

```python
df_qtde_funcionarios.plot(kind='bar', color="purple")
plt.title("Quantidade de Funcionários por Departamento")
plt.xticks(rotation=360)
```

De acordo com os dados apresentados é possível inferir que a **maior quantidade de funcionários** do Departamento está alocada **Setor de Pesquisa e Desenvolvimento**, com 961 colaboadores, seguindo o Setor de Vendas, com 446 pessoas e, por fim, o Setor de Recursos Humanos, abarcando 63 funcionários.

b) Quantidade de Funcionários por Nível de Escolaridade:


```python
df_funcionarios_escolaridade = df.groupby("Education")[["EmployeeCount"]].count()
df_funcionarios_escolaridade
```

```python
df_funcionarios_escolaridade.plot(kind='line', color="blue")
plt.title("Quantidade de Funcionários por Departamento")
```

Diante dos dados exibidos acima, constata-se que o Departamento possui a **maior parte de seus funcionários com Grau Intermediário de Instrução**, o que representa cerca de 572 colaboradores. Ainda, apenas 48 pessoas apresentam o Nível Máximo Educacional.

c) Relação Amistosa no Trabalho:


```python
df_funcionarios_atrito = df.groupby("Attrition")[["EmployeeCount"]].count()
df_funcionarios_atrito
```

```python
def valor_absoluto(pct, allvals):
    absoluto = int(pct / 100. * sum(allvals))
    return f'{absoluto}'

labels = ['SIM', 'NÃO']
sizes = [237, 1233]
colors = ['blue', 'lightblue']

plt.pie(sizes, labels=labels, colors=colors, startangle=140, autopct=lambda pct: valor_absoluto(pct, sizes),
        textprops={'fontsize': 10, 'color': 'black'})
plt.title("Quantidade de Funcionários que possui ou não atrito na Empresa")

plt.show()
```

Por fim, é possível observar a partir dos dados obtidos do BD que o Departamento possui um ótimo relacionamento interpessoal com os Funcionários do Setor, visto que do total de colaboradores (1469) apenas 236 já apresentaram algum tipo de problema de relacionamento.
